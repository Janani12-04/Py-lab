```python
import numpy as np
import matplotlib.pyplot as plt
ages = np.random.randint(10, 71, size=100)
plt.figure(figsize=(6, 4))
plt.hist(ages, bins=10, color='blue', edgecolor='black')
plt.title('Histogram of Ages')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.tight_layout()
plt.show()
```
