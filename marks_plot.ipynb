```python
import numpy as np
import matplotlib.pyplot as plt
subjects = np.array(['Math', 'Science', 'English', 'History', 'Art'])
marks = np.array([88, 70, 85, 76, 90])

plt.figure(figsize=(6, 4))
plt.bar(subjects, marks, color='skyblue')
plt.title('Marks in 5 Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.tight_layout()
plt.show()
```
