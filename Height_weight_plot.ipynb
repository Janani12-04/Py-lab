```python
import numpy as np
import matplotlib.pyplot as plt
height = np.array([150, 160, 165, 170, 175, 180, 160, 170, 155, 185])
weight = np.array([50, 55, 60, 65, 70, 75, 58, 68, 54, 80])

plt.figure(figsize=(6, 4))
plt.scatter(height, weight, color='red')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
